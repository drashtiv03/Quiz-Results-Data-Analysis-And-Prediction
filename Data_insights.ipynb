I have tried to derive various parameters determining performance of a user:

1. accuracy (already in dataset)

2. percentage: final score/total score of test (derived)

3. mistake ratio: mistakes_corrected/total mistakes made previously (derived)

4. total students: an intermediate parameter to find rank ratio
 (derived)

5. rank ratio: rank of student/total num of students (derived)

---



```python
import pandas as pd
import numpy as np

# Ensure numeric columns are properly converted
df['final_score'] = pd.to_numeric(df['final_score'], errors='coerce')
df['questions_count'] = pd.to_numeric(df['questions_count'], errors='coerce')
df['better_than'] = pd.to_numeric(df['better_than'], errors='coerce')
df['rank_numeric']=pd.to_numeric(df['rank_numeric'], errors='coerce')
df['initial_mistake_count'] = pd.to_numeric(df['initial_mistake_count'], errors='coerce')
df['mistakes_corrected'] = pd.to_numeric(df['mistakes_corrected'], errors='coerce')

# 1. Calculate the percentage column
df['percentage'] = (df['final_score'] / (df['questions_count'] * 4)) * 100

# 2. Calculate the total students column and rank ratio
df['total_students'] = df['better_than'] + df['rank_numeric']
df['rank_fraction'] = df['rank_numeric'] / df['total_students']

# 3. Calculate the ratio of mistakes corrected to the initial mistake count
df['mistake_ratio'] = np.where(df['initial_mistake_count'] == 0, 0,
                               df['mistakes_corrected'] / df['initial_mistake_count'])


```

The repetition of topics have given the numbers in sorted fashion of their start and end time to display the no. of attempt for the same topic.Hence if the topic is  Body Fluids And Circulation with name 3, it means the quiz of topic  Body Fluids And Circulation has been attempted 3rd time by the same user according to start and end time.

```python
# Convert 'at_started' to datetime
df['started_at'] = pd.to_datetime(df['started_at'])

# Sort by 'at_started' within each 'topic'
df = df.sort_values(by=['topic', 'started_at'])

# Assign sequential numbers based on sorted 'topic' groups
df['name'] = df.groupby('topic').cumcount() + 1
df['name'] = df['name'].apply(lambda x: f'{x}')  # Optional: format as 'name1', 'name2', etc.

# Display the result
print(df['topic'],df['name'])
```

Effect of performance(directly proportional to 4 parameters,out of which some are directly taken while some are derived):

The 4 paramters include


*   Accuracy
*   Percentage(marks obtained/total marks)*100
*   Mistake Ratio(number of mistakes improved/number of mistakes made previously)
*   Rank ratio(rank of user/total no. of students-could also be displayed as a fraction to user but converted to float for plotting puposes.)

1.   W.r.t Topic:
Here the number beside the topic signifies the nth attempt for the same topic of test.So you can easily identify the progress over multiple tests of same topic as well their correspondance performance amongst various topics.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# List of metrics and titles
metrics = ['accuracy', 'percentage', 'mistake_ratio', 'rank_fraction']
titles = ['Accuracy', 'Percentage', 'Mistake Ratio', 'Rank Fraction']
# Set Seaborn style
sns.set(style="whitegrid")
data['topic_name'] = data['topic'] + data['name'].astype(str)
# Plot each metric as a bar graph
for metric, title in zip(metrics, titles):
    plt.figure(figsize=(12, 6))  # Increased width for better display
    sns.barplot(x='topic_name', y=metric, data=data, color='blue')
    plt.title(title, fontsize=14)
    plt.xlabel('Topic', fontsize=12)
    plt.ylabel(metric.capitalize(), fontsize=12)
    plt.xticks(rotation=45, ha='right', fontsize=10)  # Rotate and align labels to the right
    plt.tight_layout()  # Adjust layout for better display
    plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = pd.read_csv('/content/sorted_data.csv')

# List of metrics and titles (excluding accuracy as y)
metrics = ['percentage', 'mistake_ratio', 'rank_fraction']
titles = ['Percentage', 'Mistake Ratio', 'Rank Fraction']

# Set Seaborn style
sns.set(style="whitegrid")

# Plot each metric with accuracy as the x-axis
for metric, title in zip(metrics, titles):
    plt.figure(figsize=(12, 6))  # Increased width for better display
    sns.barplot(x='accuracy', y=metric, data=data, color='blue')
    plt.title(title, fontsize=14)
    plt.xlabel('Accuracy (%)', fontsize=12)
    plt.ylabel(title, fontsize=12)
    plt.xticks(rotation=45, ha='right', fontsize=10)  # Rotate and align labels to the right
    plt.tight_layout()  # Adjust layout for better display
    plt.show()
```
